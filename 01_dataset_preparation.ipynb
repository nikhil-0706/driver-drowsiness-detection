```python
import os
import shutil
import cv2
import matplotlib.pyplot as plt
```

```python
OPEN_FOLDER = "mrleyedataset/Open-Eyes"
CLOSED_FOLDER = "mrleyedataset/Close-Eyes"

os.makedirs(OPEN_FOLDER, exist_ok=True)
os.makedirs(CLOSED_FOLDER, exist_ok=True)
```

```python
#Count images
open_images = os.listdir(OPEN_FOLDER)
closed_images = os.listdir(CLOSED_FOLDER)

print("Open eye images:", len(open_images))
print("Closed eye images:", len(closed_images))
```

```python
#Display Open Eye sample
img_path = os.path.join(OPEN_FOLDER, open_images[15])

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Open Eye")
plt.axis("off")
plt.show()
```

```python
#Display Closed Eye sample
img_path = os.path.join(CLOSED_FOLDER, closed_images[0])

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Closed Eye")
plt.axis("off")
plt.show()
```
