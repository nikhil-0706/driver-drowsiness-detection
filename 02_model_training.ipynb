```python
import os
import cv2
import pickle
import numpy as np

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
OPEN_FOLDER = "mrleyedataset/Open-Eyes"
CLOSED_FOLDER = "mrleyedataset/Close-Eyes"

IMG_SIZE = 64
```

```python
data = []
labels = []

print(type(data))
print(type(labels))
```

```python
# Load open-eye images
LIMIT = 3000

open_images = os.listdir(OPEN_FOLDER)[:LIMIT]

for i, img_name in enumerate(open_images):

    img_path = os.path.join(OPEN_FOLDER, img_name)

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        continue

    resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    flattened = resized.flatten()

    data.append(flattened)
    labels.append(0)

    if i % 500 == 0:
        print("Loaded open:", i)

print("closed loaded")
   
```

```python
# Load- closed eyes
closed_images = os.listdir(CLOSED_FOLDER)[:LIMIT]

for i, img_name in enumerate(closed_images):

    img_path = os.path.join(CLOSED_FOLDER, img_name)

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        continue

    resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    flattened = resized.flatten()

    data.append(flattened)
    labels.append(1)

    if i % 500 == 0:
        print("Loaded closed:", i)

print("Closed loaded")
```

```python
# Convert to NumPy arrays
data = np.array(data)
labels = np.array(labels)

print(data.shape)
print(np.unique(labels, return_counts=True))
```

```python
# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    data,
    labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Training data:", X_train.shape)
print("Testing data:", X_test.shape)
```

```python
# Train Model
model = SVC(kernel="linear", probability=True)

model.fit(X_train, y_train)

print("Model trained successfully")
```

```python
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
print(classification_report(y_test, y_pred))
```

```python
os.makedirs("models", exist_ok=True)

with open("models/eye_state_model.pkl", "wb") as file:
    pickle.dump(model, file)

print("Model saved successfully")
```
